```python
# Importing required libraries 
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Loading my LinkedIn connections data
usecols = ['Company' , 'Position' , 'Connected On']
df = pd.read_csv('E:\\LinkedIn Material\\LinkedIn Data\\Connections.csv' , usecols = usecols)
df['Connected On'] = pd.to_datetime(df['Connected On'])
df.head()
```

```python
# Checking the total no of LinkedIn connections I have
df.shape[0]
```

```python
# Checking the data type of the dataframe features
df.dtypes
```

```python
# Checking the no of unique companies my connections work in 
len(df['Company'].unique())
```

```python
# No of connections belonging to the same company
Top_Company_Connections = df['Company'].value_counts().head(10).sort_values()
plt.figure(figsize = (10,8))
plt.barh(Top_Company_Connections.index , width = Top_Company_Connections.values , color = 'orange')
for i, v in enumerate(Top_Company_Connections):
    plt.text(v + 0.15 , i-0.1, str(v), color='green', fontweight='bold')
plt.xticks(np.arange(0,Top_Company_Connections.max()+2,2))
plt.xlabel('No of Connections',fontweight='bold' , color = 'red' , fontsize = 12)
plt.ylabel("Connections' Organization" , fontweight='bold' , color = 'red' , fontsize = 12)
plt.title('Where do my LinkedIn connections work ? (Top 10)' , fontweight='bold' , color = '#3111E1' , fontsize = 15)
plt.xticks(fontweight='bold')
plt.yticks(fontweight='bold');
```

```python
# No of connections belonging to the same designation/role
Top_Postions_Connections = df['Position'].value_counts().head(10).sort_values()
plt.figure(figsize = (10,8))
plt.barh(Top_Postions_Connections.index, width = Top_Postions_Connections.values , color = '#FFD033')
for i, v in enumerate(Top_Postions_Connections):
    plt.text(v + 0.10 , i-0.1, str(v), color='green', fontweight='bold')
plt.xticks(np.arange(0,Top_Postions_Connections.max()+2,2))
plt.xlabel('No of Connections' , fontweight='bold' , color = 'red' , fontsize = 12)
plt.ylabel("Connections' Position" , fontweight='bold' , color = 'red' , fontsize = 12)
plt.title("What positions my LinkedIn connections hold ? (Top 10)" , fontweight='bold' , color = '#3111E1' , fontsize = 15)
plt.barh(Top_Postions_Connections.index, width = Top_Postions_Connections.values , color = '#FFD033')
plt.xticks(fontweight='bold')
plt.yticks(fontweight='bold');
```

```python
# Connections added each year
Yearwise_Connections = df['Connected On'].dt.year.value_counts().sort_index()
plt.figure(figsize = (10,8))
bar_plot = plt.bar(Yearwise_Connections.index , height = Yearwise_Connections.values , color = '#B233FF');
plt.bar(Yearwise_Connections.index , height = Yearwise_Connections.values , color = '#B233FF');
for idx,rect in enumerate(bar_plot):
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2, 1.01*height,height,ha='center', color='green', fontweight='bold')
plt.xlabel('Year' , fontweight='bold', color = 'red' , fontsize = 12)
plt.ylabel('No of connections added' , fontweight='bold' , color = 'red' , fontsize = 12)
plt.title('No of people I got connected with on LinkedIn - Yearwise' , fontweight='bold' , color = '#3111E1' , fontsize = 15)
plt.xticks(fontweight='bold')
plt.yticks(fontweight='bold');
```

```python
# Which months I have added the most connections in 
Months_Connections = df['Connected On'].dt.month.value_counts().sort_index().values
Months_list = ['Jan' , 'Feb' , 'Mar' , 'Apr' , 'May' , 'Jun' , 'Jul' , 'Aug' , 'Sep' , 'Oct' , 'Nov' , 'Dec']
Monthly_Connections_Added= pd.Series(Months_Connections, index= Months_list , name = 'Month by Month Connections')
plt.figure(figsize = (10,8))
bar_plot = plt.bar(Monthly_Connections_Added.index , height = Monthly_Connections_Added.values , color = '#D57DAC')
for idx,rect in enumerate(bar_plot):
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2, 1.01*height,height,ha='center', color='green', fontweight='bold')
plt.xlabel('Month' , fontweight='bold' , color = 'red' , fontsize = 12 )
plt.ylabel('No of connections added' , fontweight='bold' , color = 'red' , fontsize = 12)
plt.title('No of people I got connected with on LinkedIn - Monthwise' , fontweight='bold' , color = '#3111E1' , fontsize = 15)
plt.xticks(fontweight='bold')
plt.yticks(fontweight='bold');
```
